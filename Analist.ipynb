```python
df['Edad'].fillna(df['Edad'].mean(), inplace=True)
df['Genero'].fillna('Unknown', inplace=True)
df['Activista'].fillna('Not Specified', inplace=True)
df['Residencia'].fillna('Unknown', inplace=True)
df['Distrito_Residencia'].fillna('Unknown', inplace=True)
df['Tipo_Lesion'].fillna('Unknown', inplace=True)
df['Municion'].fillna('Unknown', inplace=True)
df['Obervaciones'].fillna('Unknown', inplace=True)
```

```python

df['Fecha_Evento'] = pd.to_datetime(df['Fecha_Evento'])
df['Fecha_Muerte'] = pd.to_datetime(df['Fecha_Muerte'])
```

```python
object_columns = df.select_dtypes(include=['object']).columns
print("Object type columns:")
print(object_columns)
numerical_columns = df.select_dtypes(include=['int', 'float']).columns
print("\nNumerical type columns:")
print(numerical_columns)
```

```python
df['Region'] = df['Region'].replace('West Bank', 'Cisjordania').replace('gaza strip', 'Franja de Gaza')
```

```python
df['Ciudadania'] = df['Ciudadania'].replace('Palestinian', 'Palestina').replace('Jordanian', 'Jordana').replace('American', 'USA')
```

```python
df['Activista'] = df['Activista'].replace('Not Specified', 'No especificado').replace('Object of targeted killing', 'Asesinado Selectivo')
```

```python
df['AsesinadoPor'] = df['AsesinadoPor'].replace('Israeli security forces', 'Ejercito Israeli'). replace('Palestinian civilians','Civiles Palestinos').replace ('Israeli civilians','Civiles Israelitas')
```

```python
df.head()
```

```python
def classify_features(df):
    categorical_features = []
    non_categorical_features = []
    discrete_features = []
    continuous_features = []

    for column in df.columns:
        if df[column].dtype == 'object':
            if df[column].nunique() < 10:
                categorical_features.append(column)
            else:
                non_categorical_features.append(column)
        elif df[column].dtype in ['int64', 'float64']:
            if df[column].nunique() < 10:
                discrete_features.append(column)
            else:
                continuous_features.append(column)

    return categorical_features, non_categorical_features, discrete_features, continuous_features
```

```python
categorical, non_categorical, discrete, continuous = classify_features(df)
```

```python
print("Categorical Features:", categorical)
print("Non-Categorical Features:", non_categorical)
print("Discrete Features:", discrete)
print("Continuous Features:", continuous)
```
