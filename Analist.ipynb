```python
df = df.rename(columns={'name':'Nombre','date_of_event':'Fecha_Evento','age': 'Edad','citizenship':'Ciudadania','event_location':'Lugar','event_location_district':'Distrito',
                       'event_location_region':'Region','date_of_death':'Fecha_Muerte','gender':'Genero','took_part_in_the_hostilities':'Activista','place_of_residence':'Residencia',
                       'place_of_residence_district':'Distrito_Residencia','type_of_injury':'Tipo_Lesion','ammunition':'Municion',
                       'killed_by':'AsesinadoPor', 'notes':'Obervaciones'})
```

```python

df = df.drop_duplicates()

```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Edad'].fillna(df['Edad'].mean(), inplace=True)
df['Genero'].fillna('Unknown', inplace=True)
df['Activista'].fillna('Not Specified', inplace=True)
df['Residencia'].fillna('Unknown', inplace=True)
df['Distrito_Residencia'].fillna('Unknown', inplace=True)
df['Tipo_Lesion'].fillna('Unknown', inplace=True)
df['Municion'].fillna('Unknown', inplace=True)
df['Obervaciones'].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum()
```

```python

df['Fecha_Evento'] = pd.to_datetime(df['Fecha_Evento'])
df['Fecha_Muerte'] = pd.to_datetime(df['Fecha_Muerte'])
```
